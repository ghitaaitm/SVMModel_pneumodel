```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import joblib

# Global parameters
DATA_DIR = '/content/chest_xray/chest_xray'
CATEGORIES = ['NORMAL', 'PNEUMONIA']
IMG_SIZE = (128, 128)
HOG_PARAMS = dict(
    orientations=9,
    pixels_per_cell=(8, 8),
    cells_per_block=(2, 2)
)

def load_images(split='train'):
    images, labels = [], []
    for label_idx, category in enumerate(CATEGORIES):
        path = os.path.join(DATA_DIR, split, category)
        files_list = glob.glob(os.path.join(path, '*.jpeg'))
        if not files_list:
            print(f"⚠️ No images found in {path}")
            continue
        print(f"Loading {len(files_list)} images from {path}")
        for img_path in files_list:
            try:
                img = imread(img_path, as_gray=True)
                img_resized = resize(img, IMG_SIZE)
                images.append(img_resized)
                labels.append(label_idx)
            except Exception as e:
                print(f"Error processing {img_path}: {e}")
    images = np.array(images)
    labels = np.array(labels)
    if len(images) == 0:
        raise RuntimeError(f"No images loaded for split '{split}'.")
    print(f"{split.capitalize()} set: {len(images)} images loaded.")
    return images, labels

def extract_hog_features(images):
    features = []
    hog_imgs = []
    for img in images:
        fd, hog_img = hog(img,
                         orientations=HOG_PARAMS['orientations'],
                         pixels_per_cell=HOG_PARAMS['pixels_per_cell'],
                         cells_per_block=HOG_PARAMS['cells_per_block'],
                         visualize=True, channel_axis=None)
        features.append(fd)
        hog_imgs.append(hog_img)
    return np.array(features), hog_imgs

def plot_hog_samples(images, hog_images, labels, categories, n=5):
    plt.figure(figsize=(15, 6))
    for i in range(n):
        plt.subplot(2, n, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Label: {categories[labels[i]]}")
        plt.axis('off')
        plt.subplot(2, n, i+n+1)
        plt.imshow(exposure.rescale_intensity(hog_images[i], in_range=(0, 10)), cmap='gray')
        plt.title("HOG")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

def evaluate(model, X, y, categories, set_name='Test', threshold=0.5):
    proba = model.predict_proba(X)
    y_pred = (proba[:, 1] >= threshold).astype(int)
    print(f"\nPerformance on {set_name} set (threshold={threshold}):")
    print(classification_report(y, y_pred, target_names=categories, zero_division=0))
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=categories, yticklabels=categories)
    plt.title(f"Confusion Matrix - {set_name}")
    plt.show()

def predict_image(image_path, model, pca, scaler, threshold=0.5):
    img = imread(image_path, as_gray=True)
    img = resize(img, IMG_SIZE)
    fd, _ = hog(img,
                orientations=HOG_PARAMS['orientations'],
                pixels_per_cell=HOG_PARAMS['pixels_per_cell'],
                cells_per_block=HOG_PARAMS['cells_per_block'],
                visualize=True, channel_axis=None)
    fd_scaled = scaler.transform([fd])
    fd_pca = pca.transform(fd_scaled)
    proba = model.predict_proba(fd_pca)[0]
    pred = 1 if proba[1] >= threshold else 0
    return {
        'prediction': CATEGORIES[pred],
        'probability': proba[pred],
        'normal_prob': proba[0],
        'pneumonia_prob': proba[1]
    }

def main():
    # Load data
    X_train_all, y_train_all = load_images('train')
    X_test, y_test = load_images('test')

    # Stratified train/validation split
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_all, y_train_all, test_size=0.2, random_state=42, stratify=y_train_all)

    print("Extracting HOG features...")
    X_train_hog, hog_train_imgs = extract_hog_features(X_train)
    X_val_hog, _ = extract_hog_features(X_val)
    X_test_hog, _ = extract_hog_features(X_test)

    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_hog)
    X_val_scaled = scaler.transform(X_val_hog)
    X_test_scaled = scaler.transform(X_test_hog)

    # PCA for dimensionality reduction
    pca = PCA(n_components=0.95)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_val_pca = pca.transform(X_val_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    print(f"Original HOG feature length: {X_train_hog.shape[1]}")
    print(f"PCA components retained: {pca.n_components_}")

    # Hyperparameter tuning for SVM
    param_grid = {
        'C': [1, 10, 50, 100],
        'gamma': [0.0001, 0.001, 0.01],
        'class_weight': ['balanced']
    }
    svm = SVC(kernel='rbf', probability=True)
    grid = GridSearchCV(svm, param_grid, cv=3, scoring='f1_weighted', n_jobs=-1)
    grid.fit(X_train_pca, y_train)

    print("Best SVM params:", grid.best_params_)
    svm_best = grid.best_estimator_

    # Evaluate with default threshold 0.5
    evaluate(svm_best, X_train_pca, y_train, CATEGORIES, 'Train', threshold=0.5)
    evaluate(svm_best, X_val_pca, y_val, CATEGORIES, 'Validation', threshold=0.5)
    evaluate(svm_best, X_test_pca, y_test, CATEGORIES, 'Test', threshold=0.5)

    # Try lowering threshold to improve NORMAL recall
    print("\nTrying threshold = 0.3 on Validation set to improve NORMAL recall:")
    evaluate(svm_best, X_val_pca, y_val, CATEGORIES, 'Validation', threshold=0.3)

    # Save model and preprocessors
    joblib.dump({
        'svm': svm_best,
        'pca': pca,
        'scaler': scaler,
        'hog_params': HOG_PARAMS
    }, 'pneumonia_hog_svm_model.joblib')

    plot_hog_samples(X_train, hog_train_imgs, y_train, CATEGORIES, n=5)

    # Example prediction
    sample_img = os.path.join(DATA_DIR, 'test', 'PNEUMONIA', 'person1_virus_6.jpeg')
    result = predict_image(sample_img, svm_best, pca, scaler, threshold=0.3)
    print(f"\nExample prediction on {os.path.basename(sample_img)} (threshold=0.3):")
    print(f"Prediction: {result['prediction']} (Confidence: {result['probability']:.2%})")
    print(f"Details: NORMAL {result['normal_prob']:.2%}, PNEUMONIA {result['pneumonia_prob']:.2%}")

if __name__ == "__main__":
    main()
```
