```python
# Install and setup Kaggle API
!pip install kaggle
from google.colab import files
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
import seaborn as sns
import joblib

# Upload kaggle.json
print("Please upload your kaggle.json file")
uploaded = files.upload()
os.makedirs('/root/.kaggle', exist_ok=True)
os.rename('kaggle.json', '/root/.kaggle/kaggle.json')
os.chmod('/root/.kaggle/kaggle.json', 600)

# Download and unzip dataset
!kaggle datasets download -d paultimothymooney/chest-xray-pneumonia
!unzip -q chest-xray-pneumonia.zip -d /content/chest_xray

# Configuration
data_dir = '/content/chest_xray/chest_xray'
categories = ['NORMAL', 'PNEUMONIA']
img_size = (128, 128)
orientations = 9
pixels_per_cell = (8, 8)
cells_per_block = (2, 2)

# 1. Load and preprocess data
def load_and_preprocess_data(split='train'):
    images = []
    labels = []
    for category in categories:
        path = os.path.join(data_dir, split, category)
        files = glob.glob(os.path.join(path, '*.jpeg'))
        if not files:
            print(f"Warning: No images found in {path}")
        else:
            print(f"Found {len(files)} images in {path}")
        class_num = categories.index(category)
        for img_path in files:
            try:
                img = imread(img_path, as_gray=True)
                img = resize(img, img_size)
                images.append(img)
                labels.append(class_num)
            except Exception as e:
                print(f"Error processing {img_path}: {str(e)}")
                continue
    images = np.array(images)
    labels = np.array(labels)
    print(f"Loaded {len(images)} images for {split} set")
    if len(images) == 0:
        raise ValueError(f"No images loaded for {split} set")
    if len(images) != len(labels):
        raise ValueError(f"Mismatch: {len(images)} images, {len(labels)} labels")
    return images, labels

# Load data
X_train, y_train = load_and_preprocess_data('train')
X_test, y_test = load_and_preprocess_data('test')

# Verify data
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
if len(X_train) < 2:
    raise ValueError("Not enough samples for train-test split")

# Split validation from train
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# 2. Extract HOG features
def extract_hog_features(images):
    hog_features = []
    hog_images = []
    for image in images:
        fd, hog_image = hog(image, orientations=orientations,
                            pixels_per_cell=pixels_per_cell,
                            cells_per_block=cells_per_block,
                            visualize=True, channel_axis=None)
        hog_features.append(fd)
        hog_images.append(hog_image)
    return np.array(hog_features), hog_images

print("Extracting HOG features...")
X_train_hog, hog_images_train = extract_hog_features(X_train)
X_val_hog, _ = extract_hog_features(X_val)
X_test_hog, _ = extract_hog_features(X_test)

# 3. Normalize and reduce dimensionality
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_hog)
X_val_scaled = scaler.transform(X_val_hog)
X_test_scaled = scaler.transform(X_test_hog)

pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
X_test_pca = pca.transform(X_test_scaled)

print(f"Original HOG features: {X_train_hog.shape[1]}")
print(f"PCA components retained: {pca.n_components_}")

# 4. Train SVM model
svm = SVC(kernel='rbf', C=10, gamma=0.001, class_weight='balanced', probability=True)
svm.fit(X_train_pca, y_train)

# 5. Evaluate model
def evaluate_model(model, X, y, set_name):
    y_pred = model.predict(X)
    print(f"\nPerformance on {set_name}:")
    print(classification_report(y, y_pred, target_names=categories))
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=categories, yticklabels=categories)
    plt.title(f"Confusion Matrix - {set_name}")
    plt.show()

evaluate_model(svm, X_train_pca, y_train, "Train")
evaluate_model(svm, X_val_pca, y_val, "Validation")
evaluate_model(svm, X_test_pca, y_test, "Test")

# 6. Save model
joblib.dump({
    'svm': svm,
    'pca': pca,
    'scaler': scaler,
    'hog_params': {
        'orientations': orientations,
        'pixels_per_cell': pixels_per_cell,
        'cells_per_block': cells_per_block
    }
}, 'pneumonia_hog_svm_model.joblib')

# 7. Visualize HOG features
def plot_hog_examples(images, hog_images, labels, n_samples=5):
    plt.figure(figsize=(15, 10))
    for i in range(n_samples):
        plt.subplot(2, n_samples, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Label: {categories[labels[i]]}")
        plt.axis('off')
        plt.subplot(2, n_samples, i+n_samples+1)
        hog_image_rescaled = exposure.rescale_intensity(hog_images[i], in_range=(0, 10))
        plt.imshow(hog_image_rescaled, cmap='gray')
        plt.title("HOG Features")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

plot_hog_examples(X_train[:5], hog_images_train[:5], y_train[:5])

# 8. Predict on new image
def predict_pneumonia(image_path, model, pca, scaler, hog_params):
    img = imread(image_path, as_gray=True)
    img = resize(img, img_size)
    fd, _ = hog(img, orientations=hog_params['orientations'],
                pixels_per_cell=hog_params['pixels_per_cell'],
                cells_per_block=hog_params['cells_per_block'],
                visualize=True, channel_axis=None)
    fd_scaled = scaler.transform([fd])
    fd_pca = pca.transform(fd_scaled)
    proba = model.predict_proba(fd_pca)[0]
    pred = model.predict(fd_pca)[0]
    return {
        'prediction': categories[pred],
        'probability': max(proba),
        'normal_prob': proba[0],
        'pneumonia_prob': proba[1]
    }

# Example prediction
sample_image = os.path.join(data_dir, 'test', 'PNEUMONIA', 'person1_virus_6.jpeg')
result = predict_pneumonia(sample_image, svm, pca, scaler, {
    'orientations': orientations,
    'pixels_per_cell': pixels_per_cell,
    'cells_per_block': cells_per_block
})

print("\nExample prediction:")
print(f"Image: {os.path.basename(sample_image)}")
print(f"Prediction: {result['prediction']} (Confidence: {result['probability']:.2%})")
print(f"Detail: NORMAL {result['normal_prob']:.2%}, PNEUMONIA {result['pneumonia_prob']:.2%}")
```
